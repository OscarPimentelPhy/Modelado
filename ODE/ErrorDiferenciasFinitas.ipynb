# Análisis del error en el método de diferencias finitas

**Problema**

Con la aproximación centrada para la primera derivada del método de diferencias finitas, calcular la primera derivada de la siguiente función en $x=0.5$

\begin{equation}
f(x) = -0.1x^4-0.15x^3-0.5x^2-0.25x+1.2
\end{equation}

Realizar la operación con un tamaño de paso $h=1$ y luego dividir el paso progresivamente en un factor de 10 para ver como se vuelve dominante el error de redondeo. 

**Nota:** La solución exacta para la derivada es -0.9125

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
-0.1*4*3*2*(0.5)**1 - 0.15*3*2*1 
```

```python
def f(x): return -0.1*x**4 - 0.15*x**3 - 0.5*x**2 - 0.25*x + 1.2
```

```python
errores = []
hvals = []
h = 1
x = 0.5
df_teo = -0.9125
for i in range(11):
    df = (f(x+h) - f(x-h))/(2*h)
    error = np.abs(np.abs(df_teo) - np.abs(df))
    hvals.append(h)
    errores.append(error)
    h = h/10
```

```python
hvals = np.asarray(hvals)
```

```python
plt.figure(figsize=(13,9))
plt.plot(hvals,errores,'-o',label=r'$d_xf(x=0.5)$')
#plt.plot(hvals[:7],hvals[:7]**2,'--',label=r'error$\sim h^2$')
#plt.plot(hvals[5:],1e-17/(hvals[5:]),'-.',label=r'error$\sim 1/h$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('h')
plt.ylabel(r'$d_xf(x=0.5)$')
plt.legend()
plt.savefig('error.pdf')
```
